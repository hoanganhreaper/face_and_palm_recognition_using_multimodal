```python
%matplotlib inline
import sys
from pathlib import Path
import os

import matplotlib.pyplot as plt
import PIL

# Add parent directory to import autocrop module
sys.path.insert(0, Path.cwd().parent)

from autocrop import Cropper

# Function to crop and display image
def crop_and_display(image_path, output_dir):
    c = Cropper()
    img_array = c.crop(image_path)
    plt.imshow(img_array, cmap='Greys_r')
    plt.show()
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    # Get the filename without extension
    filename = os.path.splitext(os.path.basename(image_path))[0]
    # Save the cropped image to the output directory
    output_path = os.path.join(output_dir, f"{filename}_cropped.jpg")
    PIL.Image.fromarray(img_array).save(output_path)
    print("Cropped image saved to:", output_path)

# Directory containing the images
image_dir = "quyduong_frames"

# Output directory for cropped images
output_dir = "quyduong_faces"

# List all files in the directory
image_files = os.listdir(image_dir)

# Filter only image files
image_files = [file for file in image_files if file.endswith('.jpg')]

# Loop through each image file, crop, and display
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    print("Cropping image:", image_path)
    crop_and_display(image_path, output_dir)
```
